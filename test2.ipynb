```python
import matplotlib.pyplot as plt
import pandas as pd
import pymongo
```

```python
from db.history import HistoryModel

plt.style.use('_mpl-gallery')
histories1 = HistoryModel.find({ "user_id": 1})
df1 = pd.DataFrame.from_dict(histories1)
print(df1.iloc[-1]['ETH'])
print(df1.iloc[-1]['Total'])
```

```python
from db.history import HistoryModel

plt.style.use('_mpl-gallery')
histories2 = HistoryModel.find({ "user_id": 2 })
df2 = pd.DataFrame.from_dict(histories2)
print(df2.iloc[-1]['ETH'])
print(df2.iloc[-1]['Total'])
```

```python
df = pd.DataFrame({ 'ETH1': df1['ETH'], 'ETH2': df2['ETH']})
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots()
# make a plot
ax.plot(df.index,
        df['ETH1'],
        color="red", 
        marker="o")
ax.set_xlabel("Index", fontsize = 14)
ax.set_ylabel("ETH1",
              color="red",
              fontsize=14)
ax2=ax.twinx()
ax2.plot(df.index, df["ETH2"],color="blue",marker="o")
ax2.set_ylabel("ETH2",color="blue",fontsize=14)
plt.show()
```
